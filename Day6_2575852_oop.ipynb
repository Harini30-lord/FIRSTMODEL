```python
#4,bank account

class BankAccount:
    def __init__(self, account_number, account_holder_name, balance):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return True
        return False

    def withdraw(self, amount):
        if amount > 0 and self.balance >= amount:
            self.balance -= amount
            return True
        return False

class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance, interest_rate):
        super().__init__(account_number, account_holder_name, balance)
        self.interest_rate = interest_rate

    def calculate_interest(self):
        interest = self.balance * (self.interest_rate / 100)
        self.balance += interest

class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance, overdraft_limit):
        super().__init__(account_number, account_holder_name, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount > 0 and self.balance + self.overdraft_limit >= amount:
            self.balance -= amount
            return True
        return False

# Example usage:
savings_account = SavingsAccount("SA12345", "Alice", 1000, 2.5)  # Account number, name, balance, interest rate
checking_account = CheckingAccount("CA67890", "Bob", 500, 100)  # Account number, name, balance, overdraft limit

# Deposit and withdraw from savings account
savings_account.deposit(500)
savings_account.calculate_interest()
savings_account.withdraw(200)

# Deposit and withdraw from checking account
checking_account.deposit(200)
checking_account.withdraw(600)

# Display balances
print(f"Savings Account Balance: ${savings_account.balance}")
print(f"Checking Account Balance: ${checking_account.balance}")
```

```python
#encapsulation And comments

class BankAccount:
    def __init__(self, account_number, account_holder_name, balance):
        """
        Initialize a Bank Account.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float): The account balance.

        Attributes:
            account_number (str): The account number (private).
            account_holder_name (str): The account holder's name (private).
            _balance (float): The account balance (private).
        """
        self.__account_number = account_number
        self.__account_holder_name = account_holder_name
        self._balance = balance                            # Protected attribute, can be accessed directly within subclasses

    def get_account_number(self):
        return self.__account_number

    def get_account_holder_name(self):
        return self.__account_holder_name

    def get_balance(self):
        return self._balance

    def deposit(self, amount):
        """
        Deposit money into the account.

        Args:
            amount (float): The amount to deposit.

        Returns:
            bool: True if the deposit was successful, False otherwise.
        """
        if amount > 0:
            self._balance += amount
            return True
        return False

    def withdraw(self, amount):
        """
        Withdraw money from the account.

        Args:
            amount (float): The amount to withdraw.

        Returns:
            bool: True if the withdrawal was successful, False otherwise.
        """
        if amount > 0 and self._balance >= amount:
            self._balance -= amount
            return True
        return False

class SavingsAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance, interest_rate):
        """
        Initialize a Savings Account.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float): The account balance.
            interest_rate (float): The interest rate for the account.

        Attributes:
            interest_rate (float): The interest rate for the account.
        """
        super().__init__(account_number, account_holder_name, balance)
        self.__interest_rate = interest_rate       # Private attribute for interest rate

    def get_interest_rate(self):
        return self.__interest_rate

    def calculate_interest(self):
        """
        Calculate and add interest to the account balance.
        """
        interest = self._balance * (self.__interest_rate / 100)
        self._balance += interest

class CheckingAccount(BankAccount):
    def __init__(self, account_number, account_holder_name, balance, overdraft_limit):
        """
        Initialize a Checking Account.

        Args:
            account_number (str): The account number.
            account_holder_name (str): The account holder's name.
            balance (float): The account balance.
            overdraft_limit (float): The overdraft limit for the account.

        Attributes:
            overdraft_limit (float): The overdraft limit for the account.
        """
        super().__init__(account_number, account_holder_name, balance)
        self.__overdraft_limit = overdraft_limit             # Private attribute for overdraft limit

    def get_overdraft_limit(self):
        return self.__overdraft_limit

    def withdraw(self, amount):
        """
        Withdraw money from the account, allowing overdraft up to the overdraft limit.

        Args:
            amount (float): The amount to withdraw.

        Returns:
            bool: True if the withdrawal was successful, False otherwise.
        """
        if amount > 0 and self._balance + self.__overdraft_limit >= amount:
            self._balance -= amount
            return True
        return False

# Example usage:
savings_account = SavingsAccount("SA12345", "James", 1000, 2.5)
checking_account = CheckingAccount("CA67890", "Bobby", 500, 100)

# Deposit and withdraw from savings account
savings_account.deposit(500)
savings_account.calculate_interest()
savings_account.withdraw(200)

# Deposit and withdraw from checking account
checking_account.deposit(200)
checking_account.withdraw(600)

# Display account information
print("Savings Account Information:")
print(f"Account Number: {savings_account.get_account_number()}")
print(f"Account Holder Name: {savings_account.get_account_holder_name()}")
print(f"Account Balance: R.{savings_account.get_balance()}")
print(f"Interest Rate: {savings_account.get_interest_rate()}%")

print("\nChecking Account Information:")
print(f"Account Number: {checking_account.get_account_number()}")
print(f"Account Holder Name: {checking_account.get_account_holder_name()}")
print(f"Account Balance: ${checking_account.get_balance()}")
print(f"Overdraft Limit: ${checking_account.get_overdraft_limit()}")
```

```python
#error handling

class BankAccount:
    # ... (previous code remains the same)

    def deposit(self, amount):
        try:
            if amount > 0:
                self._balance += amount
                return True
            else:
                raise ValueError("Deposit amount must be greater than zero.")
        except ValueError as e:
            print(e)
            return False

    def withdraw(self, amount):
        try:
            if amount > 0 and self._balance >= amount:
                self._balance -= amount
                return True
            else:
                raise ValueError("Withdrawal amount must be greater than zero and not exceed the account balance.")
        except ValueError as e:
            print(e)
            return False

# Example usage:
savings_account = SavingsAccount("SA12345", "Alice", 1000, 2.5)
checking_account = CheckingAccount("CA67890", "Bob", 500, 100)

# Deposit and withdraw from savings account with error handling
if savings_account.deposit(500):
    print("Deposit to Savings Account successful.")
else:
    print("Deposit to Savings Account failed.")

if savings_account.withdraw(200):
    print("Withdrawal from Savings Account successful.")
else:
    print("Withdrawal from Savings Account failed.")

# Deposit and withdraw from checking account with error handling
if checking_account.deposit(200):
    print("Deposit to Checking Account successful.")
else:
    print("Deposit to Checking Account failed.")

if checking_account.withdraw(600):
    print("Withdrawal from Checking Account successful.")
else:
    print("Withdrawal from Checking Account failed.")
```

```python
#4.employee

class Employee:
    def __init__(self, name, employee_id):
        self.__name = name  # Private attribute for name
        self.__employee_id = employee_id  # Private attribute for employee ID

    def get_name(self):
        return self.__name

    def get_employee_id(self):
        return self.__employee_id

    # Setter methods can also be added.

    def calculate_salary(self):
        pass           # Placeholde (overridden) in subclasses

class FullTimeEmployee(Employee):
    def __init__(self, name, employee_id, annual_salary):
        super().__init__(name, employee_id)
        self.__annual_salary = annual_salary  # Private attribute for annual salary

    def get_annual_salary(self):
        return self.__annual_salary

    def calculate_salary(self):
        return self.__annual_salary

class PartTimeEmployee(Employee):
    def __init__(self, name, employee_id, hours_worked, hourly_rate):
        super().__init__(name, employee_id)
        self.__hours_worked = hours_worked      # Private attribute of hours worked
        self.__hourly_rate = hourly_rate        # Private attribute of hourly rate

    def get_hours_worked(self):
        return self.__hours_worked

    def get_hourly_rate(self):
        return self.__hourly_rate

    def calculate_salary(self):
        return self.__hours_worked * self.__hourly_rate

                                 # polymorphism
full_time_employee = FullTimeEmployee("William", 101, 60000)
part_time_employee = PartTimeEmployee("Harry", 102, 30, 20)

employees = [full_time_employee, part_time_employee]

for employee in employees:
    print(f"Employee Name: {employee.get_name()}")
    print(f"Employee ID: {employee.get_employee_id()}")
    if isinstance(employee, FullTimeEmployee):
        print(f"Annual Salary: RS{employee.calculate_salary()}")
    elif isinstance(employee, PartTimeEmployee):
        print(f"Salary for {employee.get_hours_worked()} hours: RS{employee.calculate_salary()}")
    print()
```

```python
#5.library

class Book:
    def __init__(self, title, author, book_id):
        self.__title = title  # Private attribute for book title
        self.__author = author  # Private attribute for author
        self.__book_id = book_id  # Private attribute for book ID
        self.__is_checked_out = False  # Private attribute to track if the book is checked out
        self.__due_date = None  # Private attribute to track due date for checked-out books

    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author

    def get_book_id(self):
        return self.__book_id

    def is_checked_out(self):
        return self.__is_checked_out

    def check_out(self, due_date):
        if not self.__is_checked_out:
            self.__is_checked_out = True
            self.__due_date = due_date
            return True
        return False

    def return_book(self):
        if self.__is_checked_out:
            self.__is_checked_out = False
            self.__due_date = None
            return True
        return False

    def is_overdue(self, current_date):
        if self.__is_checked_out and self.__due_date is not None and current_date > self.__due_date:
            return True
        return False

class Patron:
    def __init__(self, name, patron_id):
        self.__name = name  # Private attribute for patron name
        self.__patron_id = patron_id  # Private attribute for patron ID

    def get_name(self):
        return self.__name

    def get_patron_id(self):
        return self.__patron_id

class Transaction:
    def __init__(self, book, patron, transaction_date):
        self.__book = book
        self.__patron = patron
        self.__transaction_date = transaction_date

    def get_book(self):
        return self.__book

    def get_patron(self):
        return self.__patron

    def get_transaction_date(self):
        return self.__transaction_date

    def is_overdue(self, current_date):
        return self.__book.is_overdue(current_date)

# Example usage:
book1 = Book("Fiction Book 1", "Author A", 1)
book2 = Book("Non-Fiction Book 1", "Author B", 2)
patron1 = Patron("Alice", 101)
patron2 = Patron("Bob", 102)

# Patron checks out books
book1.check_out("2023-10-01")
book2.check_out("2023-09-25")

# Create transactions
transaction1 = Transaction(book1, patron1, "2023-09-20")
transaction2 = Transaction(book2, patron2, "2023-09-21")

# Check if books are overdue and notify patrons
current_date = "2023-09-30"
if book1.is_overdue(current_date):
    print(f"{patron1.get_name()}'s book is overdue.")
if book2.is_overdue(current_date):
    print(f"{patron2.get_name()}'s book is overdue.")
```

```python
#6.shopping cart

class Product:
    def __init__(self, product_id, name, price, stock_quantity):
        self.__product_id = product_id  # Private attribute for product ID
        self.__name = name  # Private attribute for product name
        self.__price = price  # Private attribute for product price
        self.__stock_quantity = stock_quantity  # Private attribute for stock quantity

    def get_product_id(self):
        return self.__product_id

    def get_name(self):
        return self.__name

    def get_price(self):
        return self.__price

    def get_stock_quantity(self):
        return self.__stock_quantity

    def update_stock(self, quantity):
        self.__stock_quantity += quantity

    def is_in_stock(self):
        return self.__stock_quantity > 0

    def __str__(self):
        return f"{self.__name} (ID: {self.__product_id}) - ${self.__price}"

class ShoppingCart:
    def __init__(self):
        self.__items = []  # Private attribute to store items in the cart

    def add_item(self, product, quantity):
        if product.is_in_stock():
            self.__items.append((product, quantity))
            product.update_stock(-quantity)
            return True
        return False

    def calculate_total(self):
        total = sum(product.get_price() * quantity for product, quantity in self.__items)
        return total

    def clear_cart(self):
        self.__items = []

    def __str__(self):
        return "\n".join([f"{product} - Quantity: {quantity}" for product, quantity in self.__items])

class Order:
    def __init__(self, order_id, cart, shipping_cost, discount=0):
        self.__order_id = order_id  # Private attribute for order ID
        self.__cart = cart
        self.__shipping_cost = shipping_cost
        self.__discount = discount

    def get_order_id(self):
        return self.__order_id

    def calculate_total_cost(self):
        subtotal = self.__cart.calculate_total()
        total_cost = subtotal + self.__shipping_cost - (subtotal * self.__discount)
        return total_cost

# Example usage:

# Create products
product1 = Product(1, "Laptop", 1000, 5)
product2 = Product(2, "T-shirt", 20, 10)
product3 = Product(3, "Smartphone", 500, 3)

# Create a shopping cart
cart = ShoppingCart()

# Add items to the cart
cart.add_item(product1, 2)
cart.add_item(product2, 3)
cart.add_item(product3, 1)

# Display the cart
print("Shopping Cart:")
print(cart)

# Place an order
order = Order(101, cart, shipping_cost=10, discount=0.1)

# Calculate and display the total cost of the order
total_cost = order.calculate_total_cost()
print(f"\nTotal Cost of Order {order.get_order_id()}: ${total_cost}")

# Clear the cart after placing the order
cart.clear_cart()

# Update stock quantity for product1 (simulate restocking)
product1.update_stock(3)

# Attempt to add an out-of-stock item to the cart
if cart.add_item(product1, 4):
    print("\nAdded product1 to the cart.")
else:
    print("\nProduct1 is out of stock and cannot be added to the cart.")

# Display the updated cart
print("\nUpdated Shopping Cart:")
print(cart)
```
